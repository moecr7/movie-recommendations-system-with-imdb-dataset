```python
#Dataframe manipulation library
import pandas as pd
#Math functions, we'll only need the sqrt function so let's import only that
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
#this code shows plots in outputs
%matplotlib inline
```

```python
#definding movies_df as movies.csv
movies_df = pd.read_csv('movies.csv')
#definding ratings_df as ratings_sample.csv
ratings_df = pd.read_csv('ratings_sample.csv')
#Head is a function that gets the first N rows of a dataframe. N's default is 5.
movies_df.head()
```

```python
#Removing the year from title and place it in new column using regex \d means a digit
#The phrase "expand=False" typically means that a certain feature, option, or data should not be expanded or revealed
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
#Removing the parentheses
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
#Removing the years from the 'title' column and replace it with empty
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '', regex=True)
#regex=True means that the string 'pattern' is treated as a regular expression rather than a literal string. This allows for more complex matching patterns
#Applying the strip function to get rid of any ending whitespace characters that may have appeared
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

```python
#Every genre is separated by a | so we simply have to call the split function on | then we have a list
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

```python
#Copying the movie dataframe into a new one since we won't need to use the genre information in our first case.
moviesWithGenres_df = movies_df.copy()

#For every row in the dataframe, iterate through the list of genres and place a 1 into the corresponding column
#After run this sell genres are will separately in each columns and when a movie have that genre that the value of that line is set to 1
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = 1
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df.head()
```

```python
#now movie_df is preprocessed and we have to going preprocessing ratings_df
ratings_df.head()
```

```python
#removing a column that we haven't need
ratings_df = ratings_df.drop('timestamp', axis=1)
ratings_df.head()
```

```python
#importing input values with ratings between 1 , 5
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
#trsnsform it to a pandas dataframe
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
#Fetching all columns of the movies_df where the movie titles match the titles of the input movies
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
#Then merging it so we can get the movieId. It's implicitly merging it by title.
inputMovies = pd.merge(inputId, inputMovies)
#Dropping information we won't use from the input dataframe 
#axis=1 means that we want drop a column
inputMovies = inputMovies.drop('genres', axis=1).drop('year', axis=1)
inputMovies
```

```python
#this code will show columns that match with user movieId
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

```python
#Resetting the index to avoid future issues 
#After run it will show reseted index that start with 1
userMovies = userMovies.reset_index(drop=True) #drop=True means drop previous index with new data
#Dropping unnecessary issues due to save memory and to avoid issues
userGenreTable = userMovies.drop('movieId', axis=1).drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1)
userGenreTable
```

```python
#to obtain the userProfile we have to dot(matrix multiply) user ratings to genres
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
userProfile
```

```python
#Now we have to set index based on movieId
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])
#And drop the unnecessary information
genreTable = genreTable.drop('movieId', axis=1).drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1)
genreTable.head()
#now we have genre table
```

```python
#Multiply the genres by the weights and then take the weighted average to make recommendationTable
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
#sorting values biggest to smallest
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
#Just a peek at the values
recommendationTable_df.head()
```

```python
#finding name of a movie based on movieId
movies_df.loc[movies_df['movieId'].isin(recommendationTable_df.head(20).keys())]
```

```python
#Fetching all columns of the movies_df where the movie titles match the titles of the input movies
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
#Then merging it so we can get the movieId. It's implicitly merging it by title.
inputMovies = pd.merge(inputId, inputMovies)
#Dropping information we won't use from the input dataframe 
#axis=1 means that we want drop a column
inputMovies = inputMovies.drop('genres', axis=1).drop('year', axis=1)
inputMovies
```

```python
#Filtering out users that have watched movies that the inputId has watched and storing it
userSubset = ratings_df[ratings_df['movieId'].isin(inputMovies['movieId'].tolist())]
userSubset.head()
```

```python
#groupby is a finction that its group each line by userId
userSubsetGroup = userSubset.groupby(['userId'])
```

```python
#when we call get_group we can entering each usersId, it shows which movies they have watched.
userSubsetGroup.get_group(50)
```

```python
# Sort the userSubsetGroup by the length of the second element in each tuple  
# using a lambda function as the key.   
# The sorting is done in descending order (from longest to shortest).
userSubsetGroup = sorted(userSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
userSubsetGroup[0:2]
```

```python
#for don't wasting time we take only first 100 users
userSubsetGroup = userSubsetGroup[0:100]
```

## To find similarity index and recommendation table we use Pearson Correlation Coefficient
![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/bd1ccc2979b0fd1c1aec96e386f686ae874f9ec0 "Pearson Correlation")

```python
# Store the Pearson Correlation in a dictionary, where the key is the user Id and the value is the coefficient
pearsonCorrelationDict = {}

# For every user group in our subset
for name, group in userSubsetGroup:
    # Let's start by sorting the input and current user group so the values aren't mixed up later on
    group = group.sort_values(by='movieId')
    inputMovies = inputMovies.sort_values(by='movieId')
    
    # Get the N for the formula
    nRatings = len(group)
    
    # Get the review scores for the movies that they both have in common
    temp_df = inputMovies[inputMovies['movieId'].isin(group['movieId'].tolist())]
    
    # And then store them in a temporary buffer variable in a list format to facilitate future calculations
    tempRatingList = temp_df['rating'].tolist()
    
    # Let's also put the current user group reviews in a list format
    tempGroupList = group['rating'].tolist()
    
    # Now let's calculate the Pearson correlation between two users, so called, x and y
    Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    # If the denominator is different than zero, then divide, else, 0 correlation.
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```

```python
#Now we got Pearson Correlation let's see the items
pearsonCorrelationDict.items()
```

```python
#Now we should transform it to a Pandas Dataframe with pearson
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index') #orient=index means the dictionary keys will be used as row indexes in the resulting DataFrame
pearsonDF.columns = ['similarityIndex'] #this add a column named 'similarityIndex' to add pearsonDF columns 
pearsonDF['userId'] = pearsonDF.index #adding pearsonDF.index to userId column
pearsonDF.index = range(len(pearsonDF)) #Assigns this sequence of integers as the new index for pearsonDF
pearsonDF.head() #let's see the head of pearson df
```

```python
pearsonDF['userId'] = pearsonDF['userId'].apply(lambda x: x[0]) #this code will delete () and , 
pearsonDF.head()
```

```python
#This sorts pearsonDF by similarityIndex column and add it to topUsers and rows with the highest similarity indices will be at the top of the DataFrame
topUsers=pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:50] # and it shows first 50 top users sorted from pearsonDF
topUsers.head()
```

```python
# merges two DataFrames, topUsers and ratings_df, based on the userId column
#Specifies the userId column to merge on in both DataFrames (topUsers and ratings_df). An inner join is performed to return only the matching rows based on userId
topUsersRating=topUsers.merge(ratings_df, left_on='userId', right_on='userId', how='inner')
topUsersRating.head()
```

```python
#Multiplies the similarity by the user's ratings
topUsersRating['weightedRating'] = topUsersRating['similarityIndex']*topUsersRating['rating']
topUsersRating.head()
```

```python
#Applies a sum to the topUsers after grouping it up by userId
tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
tempTopUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
tempTopUsersRating.head()
```

```python
#Creates an empty dataframe
recommendation_df = pd.DataFrame()
#Now we take the weighted average
recommendation_df['weighted average recommendation score'] = tempTopUsersRating['sum_weightedRating']/tempTopUsersRating['sum_similarityIndex']
recommendation_df['movieId'] = tempTopUsersRating.index
recommendation_df.head()
```

```python
# sorting recommendation df by weighted average recommendation score and it shows the highest similarity movie
recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
recommendation_df.head(10)
```
